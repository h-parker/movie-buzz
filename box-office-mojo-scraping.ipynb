```python
from bs4 import BeautifulSoup
import requests
```

```python
# Make a get request to retrieve the page
html_page = requests.get('https://www.boxofficemojo.com/movies/?id=untitledtarantino.htm') 

# Pass the page contents to beautiful soup for pars
soup = BeautifulSoup(html_page.content, 'html.parser')
```

```python
soup
```

```python
info = {}

# function to get replace multiple different substrings with one 
# particular substring in a given string
def replace_multiple(string, to_replace_list, replacement):
    for to_replace in to_replace_list:
        string = string.replace(to_replace, replacement)
    return string



findings = list(soup.find_all('b'))

info['title'] = replace_multiple(str(findings[1]), ['<b>', '</b>', '<br/>'], '')
# typically, there are multiple genres, separated by a '/', so we 
# get this string into a list of genres by splitting after replacing random characters
info['genre'] = [x for x in replace_multiple(str(findings[5]), ['<b>', '</b>'], '').split('/')]
info['rating'] = replace_multiple(str(findings[7]), ['<b>', '</b>'], '')
info['budget'] = replace_multiple(str(findings[8]), ['<b>', '</b>'], '')

info
```
